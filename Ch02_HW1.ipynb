```python
import sympy
```

```python
x = sympy.symbols('x')
x
```

```python
f= x**3 + sympy.sin(x) + sympy.cos(x)
f
```

```python
# derivatives
f.diff(x)
```

```python
# Integrals
f.integrate(x)
```

```python
import seaborn as sb
sb.set_style('whitegrid')

p= sympy.plotting.plot(f,f.diff(x),f.integrate(x),(x,-10,10), show=False)
p[0].line_color='black'
p[1].line_color='blue'
p[2].line_color='red'
p.legend=True
p.show()
```
