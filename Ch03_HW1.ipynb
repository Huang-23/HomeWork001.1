## Binomial Distribution

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

# define number of trials
N = 20
# define probability of success in each trial
p = 0.75
# create a list of intigers from 0 to 20
m = list(range(0, 21, 1))
# calculate PMF
y = stats.binom.pmf(m, N, p)
# draw a plot
plt.stem(m, y, linefmt = 'k--', markerfmt = 'ko', basefmt = " ")
```

```python
# calculate CDF
z = stats.binom.cdf(m, N, p)
# draw a plot
plt.stem(m, z, linefmt = 'k--', markerfmt = 'ko', basefmt = " ")
```

```python
import scipy

# calculate all first four moments
mean, var, skew, kurt = scipy.stats.binom.stats(N, p, moments = 'mvsk')
print('meam=', mean)
print('var=', var)
print('skew=', skew)
print('kurt=', kurt)
```

```python
entropy = scipy.stats.binom.stats(N, p, moments = 's')
print('entropy=', entropy)
```

## Uniform Distribution
